```python
from ase.data.pubchem import pubchem_atoms_search
```

```python
ase_mol = pubchem_atoms_search(smiles=train_set["input"][2])
```

```python
ase_mol
```

```python
import torchani
```

```python
energy_calculator = torchani.models.ANI2x(periodic_table_index=True).ase()
```

```python
ase_mol.set_calculator(energy_calculator)
```

```python
from ase.optimize.precon import PreconLBFGS

opt = PreconLBFGS(ase_mol)
opt.run(fmax=0.01)
```

```python
nglview.show_ase(ase_mol)
```

```python
import ase
from ase.md.velocitydistribution import MaxwellBoltzmannDistribution
from ase.md.verlet import VelocityVerlet
from ase import units
from ase.md.langevin import Langevin
from ase.io.trajectory import Trajectory
```

```python
# this initialises the velocities by drawing from the appropriate distribution at T=300K
MaxwellBoltzmannDistribution(ase_mol, temperature_K=300) 

# We now create a molecular dynamics object that can be used to run Langevin dynamics
# the parameters after the structure are the time step, the temperature, and the friction constant (in units of picoseconds)
dynamics = Langevin(ase_mol, temperature_K=300, timestep=0.5*units.fs, friction=0.01) 


# While the dynamics is running, we want to collect some data! This is achieved by creating a function to the dynamics object
# which gets called after some number of steps, and it can report to us what is happenning to the molecule, record its trajectory, etc. 
xyzfile = open('ase_mol300.xyz', 'w') # the file we are going to record the structures to, the visualiser application "Ovito" can read such XYZ files. 
def report():
    print("Time: {:.3f} fs  |  Potential Energy: {:.3f} eV  |  Kinetic Energy: {:.3f} K".format(dynamics.get_time()/units.fs, ase_mol.get_potential_energy(),
        ase_mol.get_kinetic_energy()/(len(ase_mol)*(3.0/2.0)*units.kB)))
    ase.io.write(xyzfile, ase_mol, format="extxyz")

# notice how we print the kinetic energy in units of Kelvin, but it is not the thermodynamic temperature (which is kept constant)

dynamics.attach(report, interval=1) # the interval argument specifies how many steps of dynamics are run between each call to record the trajectory
dynamics.run(1000)
xyzfile.close()
del dynamics
```
